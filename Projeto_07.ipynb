```python
import pandas as pd
import numpy as np
import torch.nn as nn
from skorch import NeuralNetClassifier
import torch
from sklearn.model_selection import cross_val_score
```

###Etapa 02: Base de dados

```python
np.random.seed(123)
torch.manual_seed(123)
```

```python
base = pd.read_csv('/content/iris.csv')
previsores = base.iloc[:, 0:4].values
classe = base.iloc[:, 4].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
classe = encoder.fit_transform(classe)
```

```python
np.unique(classe)
```

```python
# Mudando o tipo
previsores = previsores.astype('float32')
classe = classe.astype('int64')
```

###Etapa 03: Construção do Modelo

```python
class classificador_torch(nn.Module):
    def __init__(self):
        super().__init__()
        self.dense0 = nn.Linear(4, 16)
        self.activation0 = nn.ReLU()
        self.dense1 = nn.Linear(16, 16)
        self.activation1 = nn.ReLU()
        self.dense2 = nn.Linear(16, 3)

    def forward(self, X):
        X = self.dense0(X)
        X = self.activation0(X)
        X = self.dense1(X)
        X = self.activation1(X)
        X = self.dense2(X)
        return X
```

```python
classificador_sklearn = NeuralNetClassifier(module = classificador_torch,
                                            criterion = torch.nn.CrossEntropyLoss,
                                            max_epochs = 1000,
                                            batch_size = 10,
                                            train_split = False)
```

###Etapa 04: Validação Cruzada

```python
resultados = cross_val_score(classificador_sklearn, previsores, classe, cv = 5,
                             scoring = 'accuracy')
```

```python
media = resultados.mean()
desvio = resultados.std()
```
